# LassoNet

```python
import numpy as np
from sklearn.datasets import load_breast_cancer, load_diabetes, load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

## 1. Breast Cancer Dataset (Classification)
- 569 samples, 30 features
- Binary classification

```python
cancer = load_breast_cancer()
X_cancer, y_cancer = cancer.data, cancer.target

print(f"X shape: {X_cancer.shape}")
print(f"y shape: {y_cancer.shape}")
print(f"Feature names: {cancer.feature_names[:5]}...")  # 5 features đầu
print(f"Classes: {cancer.target_names}")
print(f"Class distribution: {np.bincount(y_cancer)}")
```

## 2. Diabetes Dataset (Regression)
- 442 samples, 10 features
- Predict disease progression

```python
diabetes = load_diabetes()
X_diabetes, y_diabetes = diabetes.data, diabetes.target

print(f"X shape: {X_diabetes.shape}")
print(f"y shape: {y_diabetes.shape}")
print(f"Feature names: {diabetes.feature_names}")
print(f"y range: [{y_diabetes.min():.1f}, {y_diabetes.max():.1f}]")
```

## 3. Iris Dataset (Multi-class Classification)

```python
iris = load_iris()
X_iris, y_iris = iris.data, iris.target

print(f"X shape: {X_iris.shape}")
print(f"y shape: {y_iris.shape}")
print(f"Feature names: {iris.feature_names}")
print(f"Classes: {iris.target_names}")
print(f"Class distribution: {np.bincount(y_iris)}")
```
